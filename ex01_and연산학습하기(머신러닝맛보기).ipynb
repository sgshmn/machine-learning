환경설정

```python
# 필요한 도구 import
import pandas as pd
# 머신러닝 모델 - 최근접 이웃 분류 모델
from sklearn.neighbors import KNeighborsClassifier
# 모델 평가
from sklearn.metrics import accuracy_score
```

데이터 준비

```python
A = [0, 0, 1, 1, 1, 0, 1, 0]
B = [0, 1, 0, 1, 0, 0, 1, 1]
AandB = [0, 0, 0, 1, 0, 0, 1, 0]

dict1 = {'A': A, 'B': B, 'A and B': AandB}
df1 = pd.DataFrame(dict1)
df1
```

데이터 전처리
 - 문제, 답 분리
 - 훈련용, 테스트용 분리

```python
# 데이터 접근(인덱싱, 슬라이싱)
X = df1[['A', 'B']]# A, B
# X = df1.loc[:,'A':'B']
# X = df1.loc[:,['A', 'B']]

y =  df1['A and B']# A and B


display(X) # 출력문 - 기존 데이터 형식을 그대로 출력한다
display(y) # 출력문
```

```python
# train 분리
X_train = X.iloc[:6]
y_train = y.iloc[:6]

# test 분리
X_test = X.iloc[6:]
y_test = y.iloc[6:]

print('훈련용문제:', X_train.shape)
print('훈련용답:', y_train.shape)
print('테스트용문제:', X_test.shape)
print('테스트용답:', y_test.shape)
```

데이터 탐색(살펴보기 eda)(탐색적 데이터 분석, 우리 단계에서는 아직 필요 없다)exploratory data analysis

```python
y_train
# 클래스 = 분류할 범주 = Label
# 클래스 개수 2개
# shape 배열의 크기
# size 데이터의 개수
print(y_train.unique().size)
```

ML 모델 학습하기(모델링)

```python
# 모델 생성
knn_model = KNeighborsClassifier(n_neighbors=1) 
```

```python
# 모델 학습
knn_model.fit(X_train, y_train)
```

```python
# 모델 예측
# 1. 가장 가까운 거리에 있는 훈련데이터 찾기 k = 1
# 2. 정답을 확인하고 test에 대해서 똑같이(더 많은 쪽) 출력

pre = knn_model.predict(X_test)
pre
```

```python
y_test
```

```python
# 모델 평가
knn_model.score(X_test, y_test) # 정확도 지표
# 1이면 다 맞췄다는 뜻
```

```python
# accuracy_score 정확도 지표
accuracy_score(y_test, pre)
```
