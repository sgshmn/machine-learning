```python
# apply 함수 사용해서 Age 결측치 처리 자동화
# apply : 함수를 연결하는 함수


# apply 예시
# 사용자 함수 정의
# female -> FEMALE

# def change_name(data):
#     if data[0] == 'f':
#         return 'FEMALE'
#     return data
    
# train['Sex'].apply(change_name)


def change_name2(data):
    if data['Sex'] == 'female':
        return 'FEMALE'
    return data
    
train[['Sex', 'Fare']].apply(change_name2, axis=1)
```

```python
# 데이터프레임은 방향이 2개다
# 행 방향(열 단위 처리)
# 열 방향(행 단위 처리)
train[['Sex', 'Fare']]
```

```python
# 새로운 함수
# numpy
# np.isnan()
a = np.array([1,2,np.nan])
np.isnan(a)
pd.isnull(a)
```

```python
# 나이 결측치 처리하는 함수 정의
def fill_age(row):  # 행 단위로 열방향 axis=1
    # 만약 나이가 결측치라면 gb1에서 값을 반환하고
    if np.isnan(row['Age']):
        return gb1.loc[(row['Pclass'], row['Sex'])]
    else:
        return row['Age']
    # 값이 있으면 원래 나이를 반환하자
```

```python
# apply 연결해서 결측치 처리하기
# train, test
train['Age'] = train.apply(fill_age, axis=1).astype('int64')
train['Age'].isnull().sum()
```

```python
# test 도 적용
# fill_age함수 그대로 사용
test['Age'] = test.apply(fill_age, axis=1).astype('int64')
test['Age'].isnull().sum()
```

#### 가족 수
- 1 : alone
- 2 ~ 4 : small
- 5 ~ 11 : large
- train, test 모두 적용

```python
# pd.cut()
# 데이터를 특정한 구간으로 나누는 함수

# x : 데이터
# bins : 구간 설정
# labels : 구간 이름 설정

b= [0, 1, 4, 11]  # bins 구간 설정 (0, 1], (1, 4], (4, 11]  () 포함x, [] 포함
lb = ['alone', 'small', 'large']  # labels 구간 이름 설정

train['FamilyGroup'] = pd.cut(train['FamilySize'], bins=b, labels=lb)
```

```python
X_train = train.iloc[:, 1:]    # 훈련용 문제 Survived를 제외한 나머지 열
# drop을 써도 된다
y_train = train['Survived']   # 훈련용 답

X_test = test # 테스트 문제 여기서는 테스트 답을 줄 수 없다 캐글에 있음

print("X_train:", X_train.shape)
print("y_train:", y_train.shape)
print("X_test:", X_test.shape)
print("y_test는 캐글에 있음")
```

#### 특성 선택하고 one-hot encoding
- 범주형 데이터 인코딩

```python
X_train.columns # 'Sex', 'Embarked', 'Deck', 'FamilyGroup', 'Title'
cat_features = ['Sex', 'Embarked', 'Deck', 'FamilyGroup', 'Title']
cat_features
```

```python
pd.get_dummies(X_train['Sex'], prefix='Sex') # prefix 접두사를 붙여준다
```

```python
# 인코딩 자동화
# 각 컬럼에 한번씩 접근해서 인코딩
# 1 컬럼 접근 후 인코딩
# 2 기존 데이터 연결
# 3 원핫인코딩하고 기존 열 삭제

for cat_name in cat_features:
    dummy = pd.get_dummies(X_train[cat_name], prefix=cat_name)
    X_train = pd.concat([X_train, dummy], axis=1)
    X_train.drop(cat_name, axis=1, inplace=True)
```

```python
X_train.info()
X_train.head()
```

```python
for cat_name in cat_features:
    dummy = pd.get_dummies(X_test[cat_name], prefix=cat_name)
    X_test = pd.concat([X_test, dummy], axis=1)
    X_test.drop(cat_name, axis=1, inplace=True)
```

```python
X_test.info()
X_test.head()
```

```python
# Deck T가 테스트에 없음
# 쉽게 열이름 비교하는 방법

# set : 집합
# 1 중복을 허용하지 않는다
# 2 순서가 없다

# 자료형의 중복을 제거하는 필터 역할로 사용하기도 한다
# 합집합 교집합 등 집합연산을 할 수 있다
set(X_train.columns) - set(X_test.columns)
```

```python
# test에 Deck_T 열 추가
X_test['Deck_T'] = 0
```

```python
print('X_train:', X_train.shape)
print('X_test:', X_test.shape)
```

```python
X_test.info()
```

```python
# columns 키워드로 특성 순서 정리
X_test = X_test[X_train.columns]
```

#### 모델 선택과 학습

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

# 교차검증
from sklearn.model_selection import cross_val_score
```

```python
# tree model
tree_model = DecisionTreeClassifier(max_depth=5) # 최대 깊이 max_depth

# 교차검증 일반화 성능확인
tree_result = cross_val_score(tree_model, X_train, y_train, cv=5)
print('교차검증 결과:', tree_result)
print('교차검증 결과 평균:', tree_result.mean())
```

```python
# knn model1
knn_model = KNeighborsClassifier(n_neighbors=3)

# 교차검증 일반화 성능 확인
knn_result = cross_val_score(knn_model, X_train, y_train, cv=5)
print('교차검증 결과:', knn_result)
print('교차검증 결과 평균:', knn_result.mean())
```

knn모델 scaler 적용

```python
# 각 열의 데이터 범위가 다르기 때문에 비슷하게 맞춰준다
from sklearn.preprocessing import StandardScaler
```

```python
# 스케일러 생성
scaler = StandardScaler()

# 현재 데이터의 분포 범위 확인
scaler.fit(X_train)

# 데이터 변환
tf_X_train = scaler.transform(X_train)
tf_X_test = scaler.transform(X_test)
```

```python
# tf_X_train으로 교차검증 평균 확인

knn_result2 = cross_val_score(knn_model, tf_X_train, y_train, cv=5)
print('교차검증 결과:', knn_result2)
print('교차검증 결과 평균:', knn_result2.mean())
```

### 모델 평가와 예측
